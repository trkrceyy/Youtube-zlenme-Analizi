```python
#histogram
plt.figure(figsize=(8,5))
views = df['views'].dropna()
plt.hist(np.log1p(views),bins=60)
plt.title('Görüntülenme Sayısı Dağılımı(log(1+views))')
plt.xlabel('log(1+views)')
plt.ylabel('Sayı')
plt.tight_layout()
plt.savefig(os.path.join(output_dir,'fig_views_hist_log.png'))
plt.show()
```

```python
# Kategori bazlı ortalama görüntülenme ve video sayısı (top 20)
if 'category_id' in df.columns:
    cat_stats = df.groupby('category_id').agg(
        mean_views=('views','mean'),
        count_videos=('video_id','nunique')
    ).sort_values('count_videos', ascending=False)
    top_cats = cat_stats.head(20)
    plt.figure(figsize=(10,6))
    plt.bar(top_cats.index.astype(str), top_cats['mean_views'])
    plt.title('Kategori Başına Ortalama Görüntülenme (Top 20 kategori by video count)')
    plt.xlabel('category_id')
    plt.ylabel('Ortalama görüntülenme')
    plt.xticks(rotation=45)
    plt.bar(top_cats.index.astype(str), top_cats['mean_views'], color='orange')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'fig_cat_mean_views.png'))
    plt.show()
```

```python
 #Korelasyon matrisi (views, likes, dislikes, comment_count, like_ratio)
num_cols = ['views','likes','dislikes','comment_count','like_ratio','comment_ratio']
present_num_cols = [c for c in num_cols if c in df.columns]
if len(present_num_cols) >= 2:
    corr = df[present_num_cols].corr()
    plt.figure(figsize=(8,6))
    plt.imshow(corr.values)
    plt.title('Korelasyon Matrisi')
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
    plt.yticks(range(len(corr.index)), corr.index)
    plt.colorbar()
    for (i,j), val in np.ndenumerate(corr.values):
        plt.text(j, i, f"{val:.2f}", ha='center', va='center', fontsize=8, color='white' if abs(val)>0.5 else 'black')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'fig_corr_matrix.png'))
    plt.show()
```
